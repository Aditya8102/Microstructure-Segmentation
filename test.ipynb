```python
import tensorflow as tf
import numpy as np
import cv2
def normalize_and_resize(img_loc):
    # With Google Colab
    # my_img = tf.io.read_file(img_loc)

    # Read in image
    my_img = tf.io.read_file(img_loc)

    # Turn file into a tensor
    my_img = tf.image.decode_image(my_img)

    # Resize image
    my_img = tf.image.resize(my_img, size=[256, 256])

    # Normalize data
    my_img = my_img / 255

    return my_img
# Load the saved model
model_path ="resnet_eutectic"
loaded_model = tf.keras.models.load_model(model_path)
```

```python
from matplotlib import pyplot as plt 
image_path = "/Users/maniksinghsarmaal/Downloads/images.jpeg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)

```

```python
# Make predictions on new data
class_names = ['Eutectic','Non-Eutectic']
custom_image = normalize_and_resize(image_path)
print("INITIALISING MODEL FOR THE INPUT IMAGE.....")
prediction = loaded_model.predict(tf.expand_dims(custom_image, axis=0))
print(f'The class of the feed input image is : {class_names[np.argmax(prediction)]}')
```
